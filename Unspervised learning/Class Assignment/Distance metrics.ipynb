```python
import numpy as  np 
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.metrics.pairwise import euclidean_distances, manhattan_distances, pairwise_distances
```

```python
df = pd.read_csv(r"C:\Users\Suyash Tambe\Desktop\Unspervised learning\Class Assignment\Iris.csv")
df.head()
```

```python
column1 = df['SepalLengthCm']
column2 = df['SepalWidthCm']


column1_array = np.array(column1).reshape(-1, 1)
column2_array = np.array(column2).reshape(-1, 1)
```

```python
euclidean_dist = euclidean_distances(column1_array, column2_array)
```

```python
print("Euclidean distance between column1 and column2:", euclidean_dist[0][0])
# The Euclidean distance of 1.6 suggests that there is a moderate difference between the values in column1 and column2. This means that the values in these columns are not very similar but also not drastically different.
```

```python
manhattan_dist = manhattan_distances(column1_array, column2_array)
```

```python
print("Manhattan distance between column1 and column2:", manhattan_dist[0][0])
# The Manhattan distance of approximately 1.6 is very close to the Euclidean distance of 1.6. This suggests that the difference between the values in column1 and column2 is consistent across all dimensions (features), without the influence of the squared terms as in the Euclidean distance.
```

```python
minkowski_dist = pairwise_distances(column1_array, column2_array, metric='minkowski', p=3)
```

```python
print("Minkowski distance (p=3) between column1 and column2:", minkowski_dist[0][0])
#  The Minkowski distance withp=3 closely resembles the Manhattan distance, suggests that the discrepancy between the values in column1 and column2 remains consistent across all dimensions, mirroring the behavior observed with the Manhattan distance.
```

```python
from sklearn.metrics import jaccard_score
```

```python
df = pd.read_csv(r"C:\Users\Suyash Tambe\Desktop\Unspervised learning\Class Assignment\apple_quality.csv")
df
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
df
```

```python
df
```

```python
column1 = df['Quality']
column2 = df['Crunchiness']
```

```python
column1_binary = (column1 == column1.iloc[0]).astype(int)
column2_binary = (column2 == column2.iloc[0]).astype(int)
# the boolean results (good or bad) are converted to integers using .astype(int)
```

```python
jaccard_distance = 1 - jaccard_score(column1_binary, column2_binary)
```

```python
print(f"Jaccard Distance: {jaccard_distance}")
# The Jaccard Distance indicates a very high dissimilarity between the binary representations of column1 and column2.
# This suggests that the majority of values in one column do not match the corresponding values in the other column.
```

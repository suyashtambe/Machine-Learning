```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv(r"C:\Users\Suyash Tambe\Desktop\Supervised ML\Assignment 8\Naive bayes dataset.csv")
```

```python
df
```

```python
print("Dataset Information:")
print(df.info())
print("\nDataset Description:")
print(df.describe())
print("\nFirst few rows of the dataset:")
print(df.head())
```

```python
data_encoded = pd.get_dummies(df)
```

```python
# Inspect column names of the encoded dataset
print(data_encoded.columns)
```

```python
X = data_encoded.drop(' <=50K_ >50K', axis=1)  # Features
y = data_encoded[' <=50K_ >50K']  # Target variable
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
naive_bayes = GaussianNB()
naive_bayes.fit(x_train, y_train)
```

```python
y_pred = naive_bayes.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
